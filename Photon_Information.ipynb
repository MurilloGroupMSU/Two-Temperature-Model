# Get Intensity From Absorption Integral over Line of Sight

```python
class Measurements():
    
    def __init__(self, Z, A, r_array, n_e_array, n_i_array, Te_array, Ti_array, R_max = 100e-6, Nx = 100, Nz=500):
        self.Z, self.A = Z, A
        self.r_array  = r_array
        self.n_e_array = n_e_array
        self.n_i_array = n_i_array
        self.Te_array  = Te_array
        self.Ti_array  = Ti_array
        self.Zbar_array= self.n_e_array/self.n_i_array
        
        self.m_i = A*m_p
        
        self.x, self.z = np.linspace(-R_max,R_max, num=Nx), np.linspace(-R_max,R_max, num=Nz)
        self.dx, self.dz = self.x[1]-self.x[0], self.z[1]-self.z[0]
        self.X, self.Z = np.meshgrid(self.x,self.z, indexing='ij')
        self._grid = np.sqrt(self.X**2 + self.Z**2)
        
        self.make_parameter_grids()
        self.make_eff_parameters()
        self.make_spectral_parameters()
        self.fit_Te_with_spectral_Intensity()

    def convert_r_array_to_grid(self, array, fill_value='extrapolate'):
        array_func = interp1d(self.r_array, array, bounds_error=False, fill_value=fill_value)
        grid = array_func(self._grid)
        return grid

    def make_parameter_grids(self):
        self.n_e_grid = self.convert_r_array_to_grid(self.n_e_array)
        self.n_i_grid = self.convert_r_array_to_grid(self.n_i_array)
        self.Te_grid  = self.convert_r_array_to_grid(self.Te_array)
        self.Ti_grid  = self.convert_r_array_to_grid(self.Ti_array)
        self.Zbar_grid= self.convert_r_array_to_grid(self.Zbar_array)
        
    def make_eff_grid(self):
        self._grid = np.zeros_like(self.eff_grid)
        for x_i in range(self.eff_grid.shape[0]):
            for z_i in range(self.eff_grid.shape[1]):
                self._grid[x_i, z_i] = np.exp(-np.sum(self.dz*self.eff_grid[x_i,:z_i]))
        return self._grid

    def make__grid(self):
        self._grid = np.zeros((len(self.x), len(self.z), len(self.s)))
        for x_i in range(len(self.x)):
            for z_i in range(len(self.z)):
                self._grid[x_i, z_i, :] = np.exp(-np.sum(self._grid[x_i,:z_i,:]*self.dz,axis=0))
        return self._grid

    def make_eff_parameters(self):
        self.eff_grid = np.nan_to_num(JT_GMS.effective_photon_absorption_coefficient(self.m_i, self.n_i_grid, 
                                                                                       self.n_e_grid, self.Ti_grid, 
                                                                                       self.Te_grid, self.Zbar_grid), nan=1e8)
        self.eff_grid = self.make_eff_grid() 
        self.Ieff_grid = self.Te_grid**4*self.eff_grid
        self.Ieff_unnormalized = np.array([np.sum(self.dz*self.Te_grid[x_i]**4*self.eff_grid[x_i])/1e16 for x_i in range(len(self.x))])

    def make_spectral_parameters(self):
        self.s = k_B/hbar*np.geomspace(np.min(self.Te_grid*10), np.max(self.Te_grid*100),num=200 )
        self.d = self.s[1:]-self.s[:-1]
        self.s = 2**c/self.s[::-1]
        self.d = self.s[1:]-self.s[:-1]

        self._grid = JT_GMS.photon_absorption_coefficient(self.s[np.newaxis,np.newaxis,:], self.m_i,
                                                            self.n_i_grid[:,:,np.newaxis], 
                                                            self.n_e_grid[:,:,np.newaxis], 
                                                            self.Ti_grid[:,:,np.newaxis], 
                                                            self.Te_grid[:,:,np.newaxis], 
                                                            self.Zbar_grid[:,:,np.newaxis])
        self._grid = np.nan_to_num(self._grid, nan=1e10)

        self.B_grid = JT_GMS.photon_angular_frequency_density(self.s[np.newaxis,np.newaxis,:],self.Te_grid[:,:,np.newaxis])
        self.B_grid = JT_GMS.photon_wavelength_density(self.s[np.newaxis,np.newaxis,:],self.Te_grid[:,:,np.newaxis])

        self._grid = self.make__grid() 
        self.I_grid = self.B_grid * self._grid
        self.I_grid = self.B_grid * self._grid
        # 
        self.I_unnormalized_of_r = np.array([np.sum(self.I_grid[x_i], axis=0) for x_i in range(len(self.x))])
        self.I_unnormalized = np.sum(self.I_unnormalized_of_r*self.dx, axis=0)
        self.I_unnormalized_of_r = np.sum(self.I_unnormalized_of_r[:,1:]*self.d[np.newaxis,:], axis=1)
        # 
        self.I_unnormalized_of_r = np.array([np.sum(self.I_grid[x_i], axis=0) for x_i in range(len(self.x))])
        self.I_unnormalized = np.sum(self.I_unnormalized_of_r*self.dx, axis=0)
        
        FWHM_index = np.argmin(np.abs(self.I_unnormalized_of_r-np.max(self.I_unnormalized_of_r)/2 )) 
        self.FWHM = np.abs(2*self.x[FWHM_index])

    def plot_parameter(self, parameter_grid, label=''):
        plt.figure(figsize=(10, 7))
        contour = plt.contourf(self.X*1e6, self.Z*1e6, parameter_grid, levels=100, cmap='viridis')

        cbar = plt.colorbar(contour)
        cbar.set_label(label, size=20)
        cbar.ax.tick_params(labelsize=20)

        plt.xlabel('x [m]', fontsize=20)
        plt.ylabel('z [m]', fontsize=20)

        plt.tick_params(labelsize=20)

        plt.show()
        
    def plot_emissivity_and_intensity(self,Intensity_grid):
        # Gridspec is now 2x2 with sharp width ratios
        gs = gridspec.GridSpec(2,2,height_ratios=[4,1],width_ratios=[20,1])
        fig = plt.figure(figsize=(8,6))

        # Contour plot axis
        cax = fig.add_subplot(gs[0])

        # Create the contour plot
        contour = cax.contourf(self.X*1e6, self.Z*1e6, self.eff_grid, levels=100, cmap='viridis')
        cax.set_ylabel('z (m)', fontsize=20)
        cax.tick_params(labelsize=20)

        # Intensity plot axis
        lax = fig.add_subplot(gs[2], sharex=cax)

        # Create the Intensity plot
        lax.plot(self.X[:,0]*1e6, Intensity_grid/np.max(Intensity_grid), color='b')
        lax.set_xlabel('x (m)', fontsize=20)
        lax.set_ylabel('Intensity', fontsize=20)
        lax.tick_params(labelsize=20)
        lax.set_ylim(0,1)

        props = dict(boxstyle='round', facecolor='white', alpha=0.5)
        
        textstr = "FWHM = {0:.1f} m".format(self.FWHM*1e6)
        lax.text(0.03, 0.93, textstr, transform=lax.transAxes, fontsize=12,
            verticalalignment='top', bbox=props)

        # Make a subplot for the colour bar
        bax = fig.add_subplot(gs[1])

        # Use general colour bar with specific axis given.
        cbar = plt.colorbar(contour, cax=bax)
        cbar.set_label('Emissivity', size=20)
        cbar.ax.tick_params(labelsize=20)

        plt.tight_layout()
        plt.show()
        
    def fit_Te_with_spectral_Intensity(self):
        test_Boltzmann = lambda Te: JT_GMS.photon_wavelength_density(self.s, Te)
        f_to_min = lambda Te: np.linalg.norm( self.I_unnormalized/np.max(self.I_unnormalized) - test_Boltzmann(Te)/np.max(test_Boltzmann(Te)) )
        sol = root(f_to_min, 1e4)
        self.Te_fit = sol.x
        self.spectral_intensity_fit = test_Boltzmann(self.Te_fit)

    def plot_spectral_Intensity(self):
        self.fit_Te_with_spectral_Intensity()

        plt.figure(figsize=(10, 7))
        plt.plot(self.s*1e9, self.I_unnormalized/np.max(self.I_unnormalized), label="Integrated over profile")
        plt.plot(self.s*1e9, self.spectral_intensity_fit/np.max(self.spectral_intensity_fit), label="Te={0:.1f} kK".format(self.Te_fit[0]*1e-3))

        plt.xlabel(' [nm]', fontsize=20)
        plt.ylabel('Normalized Intensity', fontsize=20)

        plt.legend(fontsize=20)
        plt.tick_params(labelsize=20)

        plt.show()
```

```python
from scipy.optimize import curve_fit

def T_fit_form(t, Te_0, Ti_0, d, ei):
    T0 = (Te_0-Ti_0)
    T0 = (Te_0+Ti_0)
    return T0/2 - d*t + T0/2*np.exp(-2*t/ei)

def T_DIHfit_form(t, Te_0, d, ei):
    Ti_0 = Ti_initial
    T0 = (Te_0-Ti_0)
    T0 = (Te_0+Ti_0)
    return T0/2 - d*t + T0/2*np.exp(-2*t/ei)

Te0_fit, Ti0_fit, d_fit, ei_fit = curve_fit(T_fit_form, Xe_data['t[ns]'], Xe_data['T[kK]'] )[0]
Te0_DIHfit, d_DIHfit, ei_DIHfit = curve_fit(T_DIHfit_form, Xe_data['t[ns]'], Xe_data['T[kK]'] )[0]
Te0_DIHfit, d_DIHfit, ei_DIHfit
```
