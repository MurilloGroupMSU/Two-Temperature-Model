```python
import numpy as np
import matplotlib.pyplot as plt

import TTM
```

```python
dens = np.logspace(18, 24)

plt.loglog(dens, TTM.Fermi_energy(dens))
plt.grid(alpha=0.15)
plt.xlabel("density [1/cc]")
plt.ylabel("$E_F$ [eV]")
```

```python
temp = 1.0

plt.semilogx(dens, TTM.chemical_potential(dens, temp))
plt.grid(alpha=0.15)
plt.xlabel("density [1/cc]")
plt.ylabel("$\mu$ [eV]")
plt.title(f"temperature {temp}eV")
```
